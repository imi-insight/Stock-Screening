```python
page = requests.get("https://tradebrains.in/nifty-50-companies-list/")
soup = BeautifulSoup(page.content, 'html.parser')
htmltable = soup.find('table', { 'class' : 'tablepress tablepress-id-93' })
```

```python
list_table = tableDataText(htmltable)
df = pd.DataFrame(list_table[1:], columns=list_table[0])
```

```python
import xlsxwriter
sectors = []
for sheet, frame in dfcomp.items():
    sector = df.loc[df['Name'] == sheet, 'Industry'].iloc[0]
    if sector not in sectors:
        sectors.append(sector)
for sector in sectors:
    writer = pd.ExcelWriter(sector+'.xlsx', engine='xlsxwriter')
    for sheet, frame in dfcomp.items():
        sector_check = df.loc[df['Name'] == sheet, 'Industry'].iloc[0]
        if sector == sector_check:
            if len(sheet) < 30:
                frame.to_excel(writer, sheet_name = sheet)
            else:
                frame.to_excel(writer, sheet_name = sheet[:25])
    writer.save()
```
